```python
from skimage import metrics
import os
import cv2
from google.colab.patches import cv2_imshow
from skimage.metrics import structural_similarity as ssim
import numpy as np
```

```python
def cal_psnr(dir, gt_dir):
  psnr_values = []
  psnr_history = {}
  for img in os.listdir(dir):
    output_img = cv2.imread(os.path.join(dir, img))
    if img.lower().endswith('.png'):
        gt_img = cv2.imread(os.path.join(gt_dir, img.replace('.png', '.jpg')))
    else:
        gt_img = cv2.imread(os.path.join(gt_dir, img))
    h = output_img.shape[0]
    w = output_img.shape[1]
    if gt_img.shape != output_img.shape:
      gt_img = cv2.resize(gt_img,(w,h))

    psnr_gt = metrics.peak_signal_noise_ratio(gt_img,output_img)
    psnr_values.append(psnr_gt)
    psnr_history[str(img)] = {
        'psnr_gt' : psnr_gt,
    }
  print("Average_psnr: ", np.mean(psnr_values))
  return psnr_history

def cal_ssim(dir,gt_dir):
  ssim_values = []
  ssim_history = {}
  for img in os.listdir(dir):
    if img.lower().endswith('.png'):
        gt_img = cv2.imread(os.path.join(gt_dir, img.replace('.png', '.jpg')))
    else:
        gt_img = cv2.imread(os.path.join(gt_dir, img))
    output_img = cv2.imread(os.path.join(dir,img))
    h = output_img.shape[0]
    w = output_img.shape[1]
    if gt_img.shape != output_img.shape:
      gt_img = cv2.resize(gt_img,(w,h))

    ssim_gt = ssim(gt_img,output_img, data_range=output_img.max() - output_img.min(), multichannel=True)
    ssim_values.append(ssim_gt)
    ssim_history[str(img)] = {
        'ssim_gt' : ssim_gt,
    }
  print("Average_ssim: ", np.mean(ssim_values))
  return ssim_history
```

```python
BOPBTL_our = '/content/drive/MyDrive/Test/output_test/restored_image'
BOPBTL_pretrained = '/content/drive/MyDrive/Test/output_test_pretrained/final_output'
data_process = '/content/drive/MyDrive/Test/data_process'
```
